# Visualize one image

```python
# Grabs all of the columns 
# Convert the all columns object into a regular list
pixel_cols = list(df.columns)

# Sanity check that it is now just a list.
print(type(pixel_cols))

# Remove the label column from the list
pixel_cols.remove('label')
```

# 2. Select and engineer our features.
1. Create our `selected_features` that is the list of the columns we are going to use as our `X` data. 
2. Define our `X` and `y` data. 
2. Train-test-split our `X` and `y` data

### 1. Create our `selected_features` that is the list of the columns we are going to use as our `X` data. 

```python
# DOING THIS AGAIN JUST FOR PRACTICE 

# Grabs all of the columns 
selected_features = df.columns

# Convert the all columns object into a regular list
selected_features = list(selected_features)

# Sanity check that it is now just a list.
print(type(selected_features))

# Remove the label column from the list
# This happnes 'in place'

selected_features.remove('label')
```

### 2. Define our `X` and `y`

```python
# 2. Define our `X` and `y` data. 

X = df[selected_features]

y = df['label']
```

### 3. Train-test-split our `X` and `y` data

```python
# 3. Train-test-split our `X` and `y` data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25)
```

____
# 3. Build and train our model
1. Initalize an empty Support Vector Classifier model.
2. Fit that model with our training data. 

```python
# 1. Initalize an empty Support Vector Classifier model.
from sklearn import svm

# Initalize our Support Vector Classifier
model = svm.SVC()
```

```python
# 2. Fit that model with our training data. 
model.fit(X_train, y_train)
```

___
# 4. Evaluate our model
1. Get a baseline accuracy score.
2. Make new predictions using our test data. 
3. Print the classification report. 
4. Plot the confusion matrix of our predicted results. 

### 1. Get a baseline accuracy score.

```python
### 1. Get and print a baseline accuracy score.
accuracy = model.score(X_test, y_test)
print("Accuracy %f" % accuracy)
```

### 2. Make new predictions using our test data. 

```python
### 2. Make new predictions using our test data. 
y_pred = model.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
print(accuracy)
```

### 3. Print the classification report. 
Use the sklearn helper fuction for this. 

```python
### 3. Print the classification report. 

print(metrics.classification_report(y_test, y_pred))
```

### 4. Plot the confusion matrix of our predicted results. 

```python
### 4. Plot the confusion matrix of our predicted results.
fig, ax = plt.subplots(figsize=(13,13))

metrics.plot_confusion_matrix(estimator=model, 
                              X=X_test, 
                              y_true=y_test, 
                              ax=ax)
```

### Looking at the confusion matrix, which two clothing items were mis-classfied with eachother the most?

```python
print('The classifier mistaked Shirt and Coat the most.')
```

___
### Build a function thats input is an unfitted model, X, and y data, and runs the whole pipeline and prints a classification report and confusion matrix. 

```python
### Build a function thats input is an empty model, X, and y data, and runs the whole pipeline and prints a classification report and confusion matrix. 

def build_and_eval_model(model, X, y, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)    
    
    model.fit(X_train, y_train)    
    y_pred = model.predict(X_test)
    print(metrics.f1_score(y_test, y_pred, average='micro'))
    
    print(metrics.classification_report(y_test, y_pred))
    
    fig, ax = plt.subplots(figsize=(13,13))
    metrics.plot_confusion_matrix(estimator=model, 
                                  X=X_test, 
                                  y_true=y_test, 
                                  ax=ax)
    return model
```

## Run LogisticRegression, RandomForest, and Multinomial Naive Bayes through the function you just built and compare the results. 
1. Which classifier did the best, and which classifier did the worst. 

```python
# LogisticRegression
lr = LogisticRegression()
build_and_eval_model(lr, X, y)
```

```python
# RandomForest
rf = RandomForestClassifier()
build_and_eval_model(rf, X, y)
```

```python
# MultinomialNB
mnb = MultinomialNB()
build_and_eval_model(mnb, X, y)
```

```python
print('RandomForest model did the best and MultinomialNB model did the worst.')
```
